```python
from pathlib import Path

p = Path()
pa = []
print(p)
for path in p.rglob("*"):
    if path.__str__().find('.styl') != -1 :
        pa.append('./'+path.__str__().replace('\\','/'))
for s in pa:
    print(s)
    with open(s,'r',encoding="utf_8") as f:
        content = f.readlines()
    flag = 0
    for a in content:
        if a.find('body') != -1:
            flag = 1
    if flag == 1:
        print('yes')
        
```
